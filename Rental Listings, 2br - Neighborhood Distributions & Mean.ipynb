# Somerville 2B Rent Distribution, Davis Square vs. Union Square (Prospect Hill)

```r
data <- read.csv("listings_joined_attributes_only.csv", stringsAsFactors=FALSE)
myvars <- names(data) %in% c("ask", "bedrooms", "HOOD_NAME") 
newdata <- data[myvars]

# number of records per neighborhood
hist(newdata[newdata$HOOD_NAME == "Prospect Hill",]$ask, xlim=c(0,4000), breaks=8, xaxt="n", xlab="Asking Price", main="Distribution of Rents for 2 Bedroom Apartments, Prospect Hill")
axis(1, at = seq(0, 4000, 500)) 
hist(newdata[newdata$HOOD_NAME == "Davis Square",]$ask, xlim=c(0,4000), breaks=8, xaxt="n", xlab="Asking Price", main="Distribution of Rents for 2 Bedroom Apartments, Davis Square")
axis(1, at = seq(0, 4000, 500)) 


svg("prospect_hill.SVG", width= 10, height=10)
hist(newdata[newdata$HOOD_NAME == "Prospect Hill",]$ask, xlim=c(0,4000), breaks=8, xaxt="n", xlab="Asking Price", main="Distribution of Rents for 2 Bedroom Apartments, Prospect Hill")
axis(1, at = seq(0, 4000, 500)) 
dev.off()

svg("davis_square.SVG", width= 10, height=10)
hist(newdata[newdata$HOOD_NAME == "Davis Square",]$ask, xlim=c(0,4000), breaks=8, xaxt="n", xlab="Asking Price", main="Distribution of Rents for 2 Bedroom Apartments, Davis Square")
axis(1, at = seq(0, 4000, 500)) 
dev.off()
```

# Boston Listings, East Boston vs. Charleston

```r
boston <- read.csv("boston_listings_joined_attributes_only.csv", stringsAsFactors=FALSE)
myvars <- names(boston) %in% c("ask", "bedrooms", "Name") 
newdata <- boston[myvars]

hist(newdata[newdata$Name == "Charlestown",]$ask, xlim=c(0,6000), breaks=10, xaxt="n", xlab="Asking Price", main="Distribution of Rents for 2 Bedroom Apartments, Charlestown")
axis(1, at = seq(0, 6000, 500)) 
hist(newdata[newdata$Name == "East Boston",]$ask, xlim=c(0,6000), breaks=5, xaxt="n", xlab="Asking Price", main="Distribution of Rents for 2 Bedroom Apartments, East Boston")
axis(1, at = seq(0, 6000, 500)) 

svg("charlestown.SVG", width= 10, height=10)
hist(newdata[newdata$Name == "Charlestown",]$ask, xlim=c(0,6000), breaks=10, xaxt="n", xlab="Asking Price", main="Distribution of Rents for 2 Bedroom Apartments, Charlestown")
axis(1, at = seq(0, 6000, 500)) 
dev.off()

svg("east_boston.SVG", width= 10, height=10)
hist(newdata[newdata$Name == "East Boston",]$ask, xlim=c(0,6000), breaks=5, xaxt="n", xlab="Asking Price", main="Distribution of Rents for 2 Bedroom Apartments, East Boston")
axis(1, at = seq(0, 6000, 500)) 
dev.off()
```
